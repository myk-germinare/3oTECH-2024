## Tabela clientes - modelo para atualizar os dados de um cliente

## Tabelas pedidos e carrinhos - criação de tabelas vazias

```python
# Definição das instruções SQL CREATE TABLE
sql_create_pedidos = """
CREATE TABLE IF NOT EXISTS pedidos (
    id_ped INTEGER,
    data_ped DATE,
    cpf_cli BIGINT,
    id_car_ped INTEGER,
    valor_ped NUMERIC,
    transp_ped VARCHAR(100),
    rastreio_ped VARCHAR(100),
    end_entrega_ped VARCHAR(250),
    ped_sem_pgto BOOLEAN,
    PRIMARY KEY (id_ped, id_car_ped, cpf_cli)
);
"""

sql_create_carrinhos = """
CREATE TABLE IF NOT EXISTS carrinhos (
    id_car INTEGER,
    prod_car INTEGER,
    qtd_prod_car INTEGER,
    desconto_car NUMERIC, 
    PRIMARY KEY (id_car, prod_car)
);
"""

# Conexão ao banco de dados
try:
    conn = psycopg.connect(**connection_params)
    print("Conexão ao banco de dados estabelecida!")
except psycopg.Error as e:
    print("Não foi possível conectar ao banco de dados:", e)
    exit(1)

# Criação de um objeto cursor usando uma instrução with
with conn.cursor() as cur:
    # Criar tabela pedidos
    try:
        cur.execute(sql_create_pedidos)
        print("Tabela 'pedidos' criada, caso ainda não existisse.")
    except psycopg.Error as e:
        print("Erro ao criar a tabela 'pedidos':", e)

    # Criar tabela carrinhos
    try:
        cur.execute(sql_create_carrinhos)
        print("Tabela 'carrinhos' criada, caso ainda não existisse.")
    except psycopg.Error as e:
        print("Erro ao criar a tabela 'carrinhos':", e)

# Faz o commit da transação
conn.commit()

# Fecha o cursor e a conexão do banco de dados
cur.close()
conn.close()

print("Conexão ao banco de dados encerrada com sucesso!")
```
